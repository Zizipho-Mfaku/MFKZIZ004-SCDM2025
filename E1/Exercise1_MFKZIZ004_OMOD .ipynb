#### (2) Can the growth curve be seperated into some phases?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\mfaku\Downloads\flynn94.csv"
Growth_data = pd.read_csv(file_path)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(Growth_data['time'], Growth_data['cells'], marker='o', linestyle='-', label='Cell count')

# Ai has assisted me with this part of the code, I asked it to help me show the different phoses that the growth curve had. This will help me explain them 
# better as I can also see the different phases. 
plt.axvspan(0, 5, color='blue', alpha=0.1, label='Lag Phase')
plt.axvspan(5, 18, color='orange', alpha=0.1, label='Exponential Phase')
plt.axvspan(18, Growth_data['time'].max(), color='red', alpha=0.1, label='Stationary Phase')

# Labels and title
plt.xlabel('Time (days)')
plt.ylabel('Number of cells ($10^6$ ml$^{-1}$)')
plt.title('Cells vs Time graph to assess phytoplankton growth')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Answer on the above graph: 
- Yes the above growth curve could be seperated into different phases.
- There is a Lag phase: This happens at days 0-5, the growth of the cells is still constant here. 
- There is a Exponential phase: This happens from day 5, till about day 18.
- There is a stationary phase, from day 18 - 25. There may be a nutrient depletion at this point. 

#### (3) How different is the growth expressed in cellular carbon from the one in cell number and chlorophyll?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r"C:\Users\mfaku\Downloads\flynn94.csv"
Growth_data = pd.read_csv(file_path)

# Check the first few rows to know the column names
print(Growth_data.head())

# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # Wide figure for 3 plots side by side

# Plot 1: Cells vs Time
axes[0].plot(Growth_data['time'], Growth_data['cells'], marker='o', linestyle='-')
axes[0].set_xlabel('Time (days)')
axes[0].set_ylabel('Number of cells ($10^6$ ml$^{-1}$)')
axes[0].set_title('Cells vs Time')
axes[0].grid(True)

# Plot 2: Chlorophyll vs Time
axes[1].plot(Growth_data['time'], Growth_data['chl-a'], marker='o', linestyle='-', color='green')
axes[1].set_xlabel('Time (days)')
axes[1].set_ylabel('Chlorophyll (ng nl$^{-1}$)')
axes[1].set_title('Chlorophyll vs Time')
axes[1].grid(True)

# Plot 3: Cellular Carbon vs Time
axes[2].plot(Growth_data['time'], Growth_data['cellC'], marker='o', linestyle='-', color='orange')
axes[2].set_xlabel('Time (days)')
axes[2].set_ylabel('Cellular Carbon (pg C cell$^{-1}$)')
axes[2].set_title('Cellular Carbon vs Time')
axes[2].grid(True)

plt.tight_layout()
plt.show()
```

#### Answer on the above graphs: 
- The growth expressed in cellular carbon differs from the one in cell number and chlorophyll because its lag phase is shorter and it ends at day 3. At about day 4-16, it has an exponential growth. There are some days that are missing data between days 17-21, there are some gaps on the plot. There also seems to be a spike in cellular carbon at days 21-22, followed by a decline at day 25.
- Growth expressed by chlorophyll, shows that the lag phase that is longer ending at day 7 and after it is followed by an exponential growth that ends at day 16/17. There are some data missing between days 16-20, this is followed by a decline from day 21 suggesting that there could be less photosynthesis (thus less chlorophyll content). 
- Lastly, growth curve with number of cells has a lag phase that ends at day 15, followed by an exponential phase that ends at day 18. Compared to the other two graphs, the exponential phase is followed by a stationary phase suggesting maximum capacity has been reached.
