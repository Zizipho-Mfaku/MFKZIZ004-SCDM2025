#### (2) What happens when you change the value of the uptake parameter?

```python
for k in [0.05, 0.1, 0.2]:
    P_t = P_analytical(t, k, K, P0)
    plt.plot(t, P_t, label=f'k\'={k}')
plt.xlabel('Time (days)')
plt.ylabel('Phytoplankton (P)')
plt.title('Effect of Uptake Parameter k\' on Growth')
plt.legend()
plt.grid(True)
plt.show()
```

#### Answer: 
As you increase the uptake parameter, phytoplankton takes up the nutrient quicker. The percentage of food consumed per day increases. They reach the levelling off which indicates nutrient depletion quicker.
