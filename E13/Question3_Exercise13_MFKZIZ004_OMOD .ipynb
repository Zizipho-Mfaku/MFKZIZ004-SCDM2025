#### (3) Using the vertical eddy viscosity shown on page 83, what would be your choice of ∆z and ∆t to ensure stability?

```python
import numpy as np
import matplotlib.pyplot as plt

# New eddy viscosity
Av = 1e-4  # m^2/s

# Using the different Δz
dz_vals = [1.0, 2.0, 5.0]  # ( meters)
dt_vals = np.arange(100, 25000, 500)  # time steps from 100 to 25000 seconds

plt.figure(figsize=(8,5))
for dz in dz_vals:
    C = Av * dt_vals / dz**2
    plt.plot(dt_vals, C, label=f'Δz = {dz} m')

# Stability threshold
plt.axhline(0.5, color='red', linestyle='--', label='Stability limit (C=0.5)')

# Plot settings
plt.title("Courant Number vs Time Step (Av = 1e-4 m²/s)")
plt.xlabel("Δt (s)")
plt.ylabel("Courant Number")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

It seems that the most stable is when dz is 5.0 with increaisng / higher time steps. But with lower time steps stability will be reached when the dz is higher.
